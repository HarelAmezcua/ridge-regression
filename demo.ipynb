```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import functions.functions as f
```

```python
dir_dataset = os.path.join(os.getcwd(), 'datasets')
dict_dataset = {1: 'df_regresion_nolineal_1', 2: 'df_regresion_nolineal_2', 3: 'df_regresion_nolineal_3'}

# Desired dataset
number_dataset = 2
path_dataset = os.path.join(dir_dataset, dict_dataset[number_dataset] + '.csv')
```

```python
data_frame = pd.read_csv(path_dataset)

# Checks if dependent variable is in data frame, assign it to y and delete it
if 'y' in data_frame.columns: 
    y = data_frame['y'].to_numpy()    
    data_frame.drop(columns='y', inplace=True)

x = data_frame.to_numpy() # Independent variables nxm
```

```python
w = f.ridge_regression(x, y, 3,0.1)
```

```python
# Plot results
y_final = f.predict(x, w)

print('w:', w)


plt.scatter(x, y, color='blue')
plt.plot(x, y_final, color='red')
plt.title('Ridge regression')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
