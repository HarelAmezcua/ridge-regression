
# Polynomial Ridge Regression Analysis
### This notebook performs polynomial ridge regression analysis
The analysis includes:
1. Loading and preprocessing the dataset
2. Performing polynomial ridge regression
3. Plotting the regression results

```python
import sys
import os
parent_dir = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(parent_dir)
import numpy as np
import pandas as pd
import src.polynomial_ridge as f
%matplotlib inline
```

```python
dir_dataset = os.path.join(parent_dir, 'datasets')
dict_dataset = {1: os.path.join(dir_dataset,'df_regresion_nolineal_1.csv'), 
                2: os.path.join(dir_dataset,'df_regresion_nolineal_2.csv'), 
                3: os.path.join(dir_dataset,'df_regresion_nolineal_3.csv'),
                4: os.path.join(dir_dataset,'temp.csv')}

# Desired dataset
number_dataset = 1
```

```python
# Load dataset
data_frame = pd.read_csv(dict_dataset[number_dataset])

# Extract dependent variable 'y' if it exists
y = data_frame.pop('y').to_numpy() if 'y' in data_frame.columns else None

# Convert remaining dataframe to numpy array for independent variables
x = data_frame.to_numpy()
```

```python
w = f.polynomial_ridge(x, y, 10,0.5)
```

```python
# Plot results
y_final = f.predict(x, w).flatten()

print(x.shape, y.shape, y_final.shape)
f.plot_regression(x.flatten(),y,y_final)
```
